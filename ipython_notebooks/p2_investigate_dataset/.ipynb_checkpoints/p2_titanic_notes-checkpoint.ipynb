```python
first_class = titanic[titanic['Pclass'] == 1]
second_class = titanic[titanic['Pclass'] == 2]
third_class = titanic[titanic['Pclass'] == 3]

def pricier(pclass,cclass1,cclass2):
    p_list = []
    for p in pclass:
        for c1 in cclass1:
            if c1 >= p:
                p_list.append(p)
            else:
                for c2 in cclass2:
                    if c2 >= p:
                        p_list.append(p)
                    break
            break
    return len(p_list)


cheaper_1st = pricier(first_class['Fare'],second_class['Fare'],third_class['Fare'])
cheaper_2nd = pricier(second_class['Fare'],third_class['Fare'],first_class['Fare'])
cheaper_3rd = pricier(third_class['Fare'],second_class['Fare'],first_class['Fare'])

print("First class tickets cheaper than any second or third class tickets:",
     "{}%".format(round((cheaper_1st / len(first_class)) * 100,2)))
print("Second class tickets cheaper than any first or third class tickets:",
     "{}%".format(round((cheaper_2nd / len(second_class)) * 100,2)))
print("Third class tickets cheaper than any first or second class tickets:",
     "{}%".format(round((cheaper_3rd / len(third_class)) * 100,2)))


print(cheaper_1st,cheaper_2nd,cheaper_3rd)
print(len(first_class),len(second_class),len(third_class))

print(first_class['Fare'].min())
print(third_class['Fare'].max())
```
