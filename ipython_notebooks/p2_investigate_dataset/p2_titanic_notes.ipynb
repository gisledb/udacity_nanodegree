```python
#Making 'Sex' numeric to be able to perform mathematical operations using the variable
titanic['is_male'] = np.repeat(np.nan,len(titanic))
titanic['is_female'] = np.repeat(np.nan,len(titanic))

titanic['is_male'][titanic['Sex'] == 'male'] = True
titanic['is_male'][titanic['Sex'] == 'female'] = False
titanic['is_female'][titanic['Sex'] == 'female'] = True
titanic['is_female'][titanic['Sex'] == 'male'] = False

display(titanic.head(5))

display(titanic['is_male'].value_counts())
display(titanic['is_female'].value_counts())
display(titanic['Sex'].value_counts())
```

```python
#['Sex']#.loc[1]

women_first_class = class_survival_gender['PassengerId'].loc[1,(0,1),'female'].sum()
women_first_class_survived = class_survival_gender['PassengerId'].loc[1,1,'female'].sum()
women_first_class_died = class_survival_gender['PassengerId'].loc[1,1,'female'].sum()
survival_pct_women_first_class = (women_first_class_survived / women_first_class) * 100

women_second_class = class_survival_gender['PassengerId'].loc[2,(0,1),'female'].sum()
women_second_class_survived = class_survival_gender['PassengerId'].loc[2,1,'female'].sum()
women_second_class_died = class_survival_gender['PassengerId'].loc[2,1,'female'].sum()
survival_pct_women_second_class = (women_second_class_survived / women_second_class) * 100

women_third_class = class_survival_gender['PassengerId'].loc[3,(0,1),'female'].sum()
women_third_class_survived = class_survival_gender['PassengerId'].loc[3,1,'female'].sum()
women_third_class_died = class_survival_gender['PassengerId'].loc[3,1,'female'].sum()
survival_pct_women_third_class = (women_third_class_survived / women_third_class) * 100

survival_gender_class = pd.Series(
    [round(survival_pct_women_first_class,2),round(survival_pct_women_second_class,2),
     round(survival_pct_women_third_class,2)]
    )
     

survival_gender_class.index = 'first_class','second_class','third_class'

survival_gender_class
#class_survival_gender['PassengerId'].loc[1].sum()

#class_survival['Sex'].loc[1]

survival_gender_class.plot(kind='bar',table=True,use_index=False)

titanic.describe()
```
