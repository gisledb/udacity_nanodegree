```python
from IPython.display import display

def namestr(obj):
    return [name for name in globals() if globals()[name] is obj]

def df_five_records(*df):
#print("{0}".format(df.head(5)))
    #adding sub-function for nicer ipython output
    
    for table in df:
        
        tab = pd.DataFrame(table)
        print('table: ',namestr(table)[0])
        display(tab.head(5))
    return
```

```python
#Looking at all tables
df_five_records(
    allstarfull,awardsplayers,batting,fielding,\
    halloffame,managershalf,pitching,schools,\
    teamsfranchises,appearances,awardssharemanagers,\
    battingpost,fieldingof,homegames,master,\
    pitchingpost,seriespost,teamshalf,awardsmanagers,\
    awardsshareplayers,collegeplaying,fieldingpost,
    managers,parks,salaries,teams
    )
```
